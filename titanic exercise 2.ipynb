```python
titanic_data = pd.read_csv('titanic.csv')
```

```python
# 2.1 Tampilkan data penumpang yg memiliki gender female
kondisi_female = titanic_data['Sex'] == 'female' # cara 1

titanic_data[kondisi_female]
```

```python
# cara 2
titanic_data[titanic_data['Sex'] == 'female']
```

```python
# case 2.2
## tampilkan data name, Age. Pclass dan Fare daei pwnumpang yg usianya diatas 40

titanic_data[titanic_data['Age']>40][['Name', 'Age', 'Pclass', 'Fare']]
```

```python
# case 2.3
# tampilkan data penumpang yg harga tiketnya diatas $50 dan berangkat dari port Southhampton
titanic_data[(titanic_data['Fare'] > 50) & (titanic_data['Embarked'] == 'S')]
```

```python
# 2.4
# Tampilkan informasi Name, Embarked, SibSp, dan Parch dari penumpang yg SibSp nya lebih dari 6 atau Parch lebih dari 2
titanic_data[(titanic_data['SibSp'] > 6) | (titanic_data['Parch'] > 2)][['Name', 'Embarked', 'SibSp', 'Parch']]
```

```python
# exercise
# 1. Tampilkan nama dan gender penumpang yg survived dan memiliki usia dibawah 25 thn
titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Age'] < 25)][['Name','Sex']]
```

```python
# 2. Tampilkan data penumpang yg survived dan berangkat dari port Queenstwon atau Souyhhampton
titanic_data[(titanic_data['Survived'] == 1) & ((titanic_data['Embarked'] == 'Q') | (titanic_data['Embarked'] == 'S'))]
```

```python
# 3. Tampilkan data penumpang yg jumlah anggota keluarganya lebih dari 7
titanic_data[(titanic_data['SibSp'] + titanic_data['Parch'] >7)] # cara 1
```

```python
## cara 2
titanic_data['FamSize'] = titanic_data['SibSp'] + titanic_data['Parch']
titanic_data [titanic_data['FamSize'] > 7]
```

```python
# 4. Tampilkan data penumpang yg tidak survive, memiliki anggota keluarga lebih dari 5 dan tidak berangkat dari Quennstown
titanic_data[(titanic_data['Survived'] == 0) & (titanic_data['FamSize'] >5) & (titanic_data['Embarked'] != 'Q')]
```
